```python
import numpy as np
```

```python
class Percep:

    def __init__(self, lr=0.1, ep=20):
        # lr = learning rate
        # ep = number of training epochs
        self.lr = lr
        self.ep = ep

        # init weights and bias
        self.w = np.zeros(2)
        self.b = 0

    def step(self, z):
        # step activation
        return 1 if z >= 0 else 0

    def pred(self, x):
        # calc score
        z = np.dot(self.w, x) + self.b
        return self.step(z)

    def train(self, X, y):

        for _ in range(self.ep):
            for i in range(len(X)):

                x = X[i]
                t = y[i]

                # prediction
                y_hat = self.pred(x)

                # error
                err = t - y_hat

                # update rule
                self.w += self.lr * err * x
                self.b += self.lr * err
```

```python
def get_data(gate):

    X = np.array([
        [0,0],
        [0,1],
        [1,0],
        [1,1]
    ])

    if gate == "AND":
        y = np.array([0,0,0,1])

    elif gate == "OR":
        y = np.array([0,1,1,1])

    elif gate == "NAND":
        y = np.array([1,1,1,0])

    elif gate == "NOR":
        y = np.array([1,0,0,0])

    elif gate == "XOR":
        y = np.array([0,1,1,0])

    return X, y
```

```python
def run_gate(g):

    X, y = get_data(g)

    p = Percep(lr=0.1, ep=20)
    p.train(X, y)

    print("\nGate:", g)
    print("Weights:", p.w)
    print("Bias:", p.b)

    print("Predictions")

    for x in X:
        print(x, "->", p.pred(x))
```

```python
gates = ["AND","OR","NAND","NOR","XOR"]

for g in gates:
    run_gate(g)
```

# 1. Effect of learning rate:
Learning rate controls how much weights change after each mistake.

a. Small learning rate: slow learning , stable convergence

b. Large learning rate: faster learning , may oscillate and fail to converge

In this experiment, a moderate value like 0.1 allows stable training for linearly separable gates.

# 2. Why Same Code Learns Different Gates:
The perceptron architecture remains the same.
Only the training data changes.

The dataset defines the relationship between inputs and outputs, so the learning rule adjusts weights and bias differently for each gate.

Thus the model behavior is determined by the data, not by changing the code.
